```python
import pandas as pd
```

```python
csv = "https://raw.githubusercontent.com/erickbytes/lofipy/main/pythonmarketer.wordpress.com_searchterms_day_08_18_2023_08_18_2023.csv"
search_terms = pd.read_csv(csv, names=["terms", "visitors"])
search_terms.info()
```

```python
print("This shows what people searched to find my Python programming blog from 2016-2023, exported from Wordpress.")
print("It is a limited snapshot but offers some insight into what people are looking for when they land on lofipython.com")
search_terms.head(25)
```

```python
print("A lot of traffic had to do with installing Python.")
install_searches = search_terms[
    search_terms.terms.astype(str).str.contains(pat="install", regex=False, case=False)
]
install_searches.head(10).style.hide()
```

```python
install_percent = install_searches.index.size / search_terms.index.size
print(f"At least {install_percent:.0%} of recorded search terms were related to installing Python, pip or a specific library.")
```

```python
other_searches = search_terms[
    ~search_terms.terms.astype(str).str.contains(pat="install", regex=False, case=False)
]
other_searches.head(10).style.hide()
```

```python
nonpython_searches = search_terms[
    ~search_terms.terms.astype(str).str.contains(pat="python", regex=False, case=False)
]
nonpython_searches.head(10).style.hide()
```

```python
pip_searches = search_terms[search_terms.terms.str.contains("pip", regex=False, case=False)]
pip_searches.head(10).style.hide()
```

```python
print("Select 'Cell' > 'Run All', then type ctrl + enter in this cell to see 25 random library searches.")
library_searches = search_terms[search_terms.terms.str.contains("lib", regex=False, case=False)]
library_searches.sample(n=25).style.hide()
```

```python
print("Select 'Cell' > 'Run All', then type ctrl + enter in this cell to see 25 random search terms.")
search_terms.sample(n=25).sort_values(by="visitors", ascending=False).style.hide()
```
